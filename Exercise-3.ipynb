### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake','Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}
test_df = pd.DataFrame(data)
test_df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(test_df['preTestScore'], test_df['age'], s = test_df['age']* 2)
plt.scatter(test_df['postTestScore'], test_df['age'], s = test_df['age']* 2)
plt.xlabel("Test scores")
plt.ylabel("Age (Years)")
plt.title("PreTest and PostTest score by Age")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(test_df[test_df['female'] == 1]['preTestScore'],test_df[test_df['female'] == 1]['age'],color='blue')
plt.scatter( test_df[test_df['female'] == 0]['preTestScore'],test_df[test_df['female'] == 0]['age'],color='green')
plt.scatter(test_df[test_df['female'] == 1]['postTestScore'], test_df[test_df['female'] == 1]['age'],color='blue')
plt.scatter(test_df[test_df['female'] == 0]['postTestScore'],test_df[test_df['female'] == 0]['age'],color='green')
plt.xlabel("Test scores")
plt.ylabel("Age (Years)")
plt.title("PreTest and PostTest score by Age")
```
